```python
from astropy.io import fits as fits
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from iminuit import Minuit
from iminuit.cost import LeastSquares

import clmm
```

## Read in the 3 projections of one clusters of the 300

```python
ff_0 = '/sps/lsst/users/ccombet/TheThreeHundred/maps/lensing/Max/snap_115/maps_0019_115_0.fits'
dat_0 = fits.open(ff_0)
kappa_0 = dat_0[0].data
shear1_0 = dat_0[1].data
shear2_0 = dat_0[2].data

ff_1 = '/sps/lsst/users/ccombet/TheThreeHundred/maps/lensing/Max/snap_115/maps_0019_115_1.fits'
dat_1 = fits.open(ff_1)
kappa_1 = dat_1[0].data
shear1_1 = dat_1[1].data
shear2_1 = dat_1[2].data

ff_2 = '/sps/lsst/users/ccombet/TheThreeHundred/maps/lensing/Max/snap_115/maps_0019_115_2.fits'
dat_2 = fits.open(ff_2)
kappa_2 = dat_2[0].data
shear1_2 = dat_2[1].data
shear2_2 = dat_2[2].data
```

```python
dat_1[0].header
```

## Visualize the $\gamma_1$, $\gamma_2$ and $\kappa$ maps

```python
ra = np.arange(-1024,1024)*dat_1[0].header['CD1_1']
dec = np.arange(-1024,1024)*dat_1[0].header['CD2_2']
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15,15))
axes[0][0].imshow(np.log(kappa_0), origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-7, vmax=0.5)
axes[0][0].set_xlabel('ra [deg]')
axes[0][0].set_ylabel('dec [deg]')
axes[0][1].imshow(shear1_0, origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-0.3, vmax=0.3)
axes[0][1].set_xlabel('ra [deg]')
axes[0][1].set_ylabel('dec [deg]')
axes[0][2].imshow(shear2_0, origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-0.3, vmax=0.3)
axes[0][2].set_xlabel('ra [deg]')
axes[0][2].set_ylabel('dec [deg]')

axes[1][0].imshow(np.log(kappa_1), origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-7, vmax=0.5)
axes[1][0].set_xlabel('ra [deg]')
axes[1][0].set_ylabel('dec [deg]')
axes[1][1].imshow(shear1_1, origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-0.3, vmax=0.3)
axes[1][1].set_xlabel('ra [deg]')
axes[1][1].set_ylabel('dec [deg]')
axes[1][2].imshow(shear2_1, origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-0.3, vmax=0.3)
axes[1][2].set_xlabel('ra [deg]')
axes[1][2].set_ylabel('dec [deg]')

axes[2][0].imshow(np.log(kappa_2), origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-7, vmax=0.5)
axes[2][0].set_xlabel('ra [deg]')
axes[2][0].set_ylabel('dec [deg]')
axes[2][1].imshow(shear1_2, origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-0.3, vmax=0.3)
axes[2][1].set_xlabel('ra [deg]')
axes[2][1].set_ylabel('dec [deg]')
axes[2][2].imshow(shear2_2, origin='lower', extent=[ra[0],  ra[-1], dec[0], dec[-1]], vmin=-0.3, vmax=0.3)
axes[2][2].set_xlabel('ra [deg]')
axes[2][2].set_ylabel('dec [deg]')


fig.tight_layout()
fig.savefig("./maps.png")
```
