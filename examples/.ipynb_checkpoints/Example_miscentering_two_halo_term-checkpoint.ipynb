```python
def P(R_off, sigma_off):
    
    return ( R_off/(sigma_off**2) ) * np.exp(-0.5 * (R_off/sigma_off)**2)

```

```python
def predict_sigma_average_mis(r, cluster_z, sigma_off, moo):
    
    Roff = np.logspace(-11,np.log10(30), 10)
    
    Theta = np.linspace(0, 2*np.pi, 10)
    
    signal_at_R = []

    for R in r:
        
        print(1)
        
        R_to_int = np.logspace(-11,np.log10(R),100)
        
        val_int_theta_fixed = []

        for j, theta in enumerate(Theta):
            
            val_int_theta_roff_fixed = []
            
            for i, roff in enumerate(Roff):
                
                Rprime = np.sqrt(R_to_int**2 + roff**2 - 2*R_to_int*roff*np.cos(theta))

                val = moo.eval_sigma(Rprime, cluster_z)*(2/(R**2))*R_to_int
                
                int_R = simps(val, R_to_int)

                val_int_theta_roff_fixed.append(simps(val, R_to_int)*P(roff, sigma_off)/(2*np.pi))
                

            val_int_theta_fixed.append(simps(val_int_theta_roff_fixed, Roff))
            
        signal_at_R.append(simps(val_int_theta_fixed, Theta))
        
    return np.array(signal_at_R)
```

```python
kk = np.logspace(-12,np.log10 (15) ,10000)
```

```python
r = np.logspace(-1, np.log10(10),10)
moo = clmm.Modeling(massdef = 'mean', delta_mdef = 200, halo_profile_model = 'nfw')
mock_cosmo = Cosmology(H0=70.0, Omega_dm0=0.27-0.045, Omega_b0=0.045, Omega_k0=0.0)
cosmo_ccl = ccl.Cosmology(
    Omega_c=0.25, Omega_b=0.05, h=0.7, A_s=2.1e-9, n_s=0.96, Neff=0, Omega_g=0)
moo.set_cosmo(mock_cosmo)
moo.set_mass(cluster_m) 
moo.set_concentration(concentration)
```

```python
kk, Pk = twoh.compute_Pk(kk, cluster_z, cosmo_ccl)
```

```python
ds_unbaised = twoh.two_halo_term_unbaised(r, cluster_z, cosmo_ccl, kk, Pk)
bais = twoh.halo_bais(logm = np.log10(cluster_m), mdef = 'matter', Delta = 200, cluster_z = cluster_z, cosmo_ccl = cosmo_ccl)
```

```python
ds_mis = mis.predict_excess_mis(r, cluster_z, moo)
```

```python
plt.loglog(r,moo.eval_sigma_excess(r,cluster_z), '-o', label = r'$\Delta\Sigma_{NFW}$')
plt.loglog(r,ds_mis, '--s', label = r'$\Delta\Sigma_{NFW, mis}$')
plt.loglog(r,ds_unbaised*bais, '--^', label = r'$\Delta\Sigma_{2h}$')
plt.plot(r, ds_unbaised*bais + ds_mis,linewidth=5, label = r'$\Delta\Sigma_{tot}$')
plt.xlabel('R [Mpc]')
plt.ylabel(r'$\Delta\Sigma_+$ ' +'$[$' + r'${\rm M}$' + r'$_\odot\;$'+ r'${\rm Mpc}$'+r'$^{-2}$'r'$]$')
plt.legend()
sys.path.append('/pbs/throng/lsst/users/cpayerne/GitForThesis/DC2Analysis/modeling')
#plt.savefig('illustration', dpi = 300)
```
