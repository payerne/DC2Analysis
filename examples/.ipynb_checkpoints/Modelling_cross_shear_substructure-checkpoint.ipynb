```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
def shear(r):
    return np.exp(-r)
```

```python
shear(0)
```

```python
def r(theta, theta_i, phi, phi_i):
    a = theta**2 + theta_i**2 - 2*theta*theta_i*np.cos(phi-phi_i)
    return np.sqrt(a)
```

```python
def phi_RT(theta, theta_i, phi, phi_i):
    a = theta*np.sin(phi) - theta_i*np.sin(phi_i)
    b = theta*np.cos(phi) - theta_i*np.cos(phi_i)
    tan = a/b
    return np.arctan(tan)
```

```python
def cross_shear(theta, theta_i, phi_i):
    res = []
    for theta_ in theta:
        def __integrand__(x):
            R = r(theta_, theta_i, x, phi_i)
            ST = shear(R)
            phi_RT_ = phi_RT(theta_, theta_i, x, phi_i)
            return -ST*np.sin(2*(x-phi_RT_))
        res.append(quad(__integrand__, 0., 2*np.pi)[0]/(2*np.pi))
    return np.array(res)
```

```python
def tangential_shear(theta, theta_i, phi_i):
    res = []
    for theta_ in theta:
        def __integrand__(x):
            R = r(theta_, theta_i, x, phi_i)
            ST = shear(R)
            phi_RT_ = phi_RT(theta_, theta_i, x, phi_i)
            return -ST*np.cos(2*(x-phi_RT_))
        res.append(quad(__integrand__, 0., 2*np.pi)[0]/(2*np.pi))
    return np.array(res)
```

```python
theta_i = np.linspace(0.1, 2, 5)
phi_i = np.linspace(0.0, 2*np.pi, 5)
theta = np.linspace(0.001, 1, 10)
plt.figure(figsize = (7,7))
for i, theta_i_ in enumerate(theta_i):
    for j, phi_j_ in enumerate(phi_i):
        y = cross_shear(theta, theta_i_, phi_j_)
        linestyle = '-'
        plt.plot(theta, abs(y), linestyle, linewidth = 2)
plt.yscale('log')
plt.xlabel(r'$\theta$', fontsize = 20)
plt.ylabel(r'$|\gamma_{\times,k}|$', fontsize = 20)
#plt.ylim(1e-22, 1e-12)
plt.legend(frameon=False, fontsize = 15)
#plt.savefig('/pbs/throng/lsst/users/cpayerne/ThesisAtCCin2p3/Images/integration.png', bbox_inches='tight', dpi=300)
```

```python
def __integrand__(phi, theta_i, phi_i, theta_):
    R = r(theta_, theta_i, phi, phi_i)
    ST = shear(R)
    phi_RT_ = phi_RT(theta_, theta_i, phi, phi_i)
    #return np.sin(2*(phi-phi_RT_))
    return (-ST), np.sin(2*(phi-phi_RT_))*(-ST)
    #return np.sin(2*(phi-phi_RT_))*(-ST)
```

```python
theta_i = np.linspace(0.1, 2, 5)
phi_i = np.linspace(0.0, 2*np.pi, 5)
theta = np.linspace(0.001, 1, 10)
phi = np.linspace(-np.pi, np.pi, 100)
fig, ax = plt.subplots(len(theta_i), len(phi_i), figsize = (9,9), sharex=True, sharey=True)
plt.subplots_adjust(wspace=0, hspace=0)
for i, theta_i_ in enumerate(theta_i):
    for j, phi_i_ in enumerate(phi_i): 
        ax[i,j].plot([],[], 'x', c ='w')#, label = f'({theta_i_:.2f},{phi_i_:.2f})')
        for k, theta_ in enumerate(theta):
            y, y_ = __integrand__(phi, theta_i_, phi_i_, theta_)
            ax[i,j].plot(phi, y, 'magenta')
            ax[i,j].plot(phi, y_, 'royalblue')
            ax[i,j].plot(phi, y_*y, 'k')
            ax[i,j].set_xlabel(r'$\phi$', fontsize = 20)
            ax[i,j].set_facecolor('lightsalmon')
            ax[i,j].tick_params(axis='both', labelsize=0)
            #ax[i,j].legend(frameon = True)
            ax[i,j].set_xlim(-np.pi, np.pi)
            ax[i,j].set_ylim(-1, 1)
#plt.savefig('/pbs/throng/lsst/users/cpayerne/ThesisAtCCin2p3/Images/integrand.png', bbox_inches='tight', dpi=300)
```
