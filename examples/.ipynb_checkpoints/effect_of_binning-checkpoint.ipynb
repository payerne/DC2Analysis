```python
import matplotlib.pyplot as plt
import numpy as np
from math import gamma
from scipy.integrate import quad
```

```python
def model(x, a): return np.exp(-a*x)
```

```python
d = 1
alpha_d = np.pi**(d/2) * gamma(d/2 + 1)
```

```python
alpha_d
```

```python
def dV_dx(x):
    return alpha_d * x**(d-1)
def binng_model(x_low, x_up, a):
    def model_dV_dx(x): return dV_dx(x)*model(x, a)
    V = quad(dV_dx, x_low, x_up)[0]
    s = quad(model_dV_dx, x_low, x_up)[0]
    return s/V
```

```python
a = 10
x_corner = np.logspace(-1, 1, 11)
mean_x_corner = np.array([np.mean([x_corner[i], x_corner[i+1]]) for i in range(len(x_corner)-1)])
x_array = np.linspace(0.1, 10, 1000)
y_model = model(x_array, a)
y_model_center = model(mean_x_corner, a)
y_binned = np.zeros(len(x_corner)-1)
for i in range(len(x_corner)-1):
    y_binned[i] = binng_model(x_corner[i], x_corner[i+1], a)
```

```python
plt.figure(figsize = (7,5))
plt.loglog(x_array, y_model, label = 'true')
plt.step(mean_x_corner, y_binned, label = 'binned', where = 'mid')
plt.legend()
```

```python
a = 1
sigma_signal = .3
x_low, x_up = 1, 10
ns = 10
def error_list(n):
    x_corner = np.logspace(np.log10(x_low), np.log10(x_up), n + 1)
    mean_x_corner = np.array([np.mean([x_corner[i], x_corner[i+1]]) for i in range(len(x_corner)-1)])
    y_binned = np.zeros(len(x_corner)-1)
    #binned vector
    for i in range(len(x_corner)-1):
        y_binned[i] = binng_model(x_corner[i], x_corner[i+1], a)
    y_true_center = model(mean_x_corner, a)
    #error:
    S = np.zeros(len(x_corner)-1)
    for i in range(len(x_corner)-1):
        S[i] = quad(dV_dx, x_corner[i], x_corner[i+1])[0]
    err = sigma_signal/np.sqrt(ns * S)
    return err, y_binned, y_true_center
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15,5))
for n in range(5, 50):
    err, y_binned, y_true_center = error_list(n)
    bias = 1/np.mean(y_true_center/y_binned)
    err_mean = np.mean(err)
    err_rel =np.mean(1-(err/y_binned))
    ax[0].scatter(n, bias, c = 'k')
    ax[1].scatter(n, err_mean, c = 'cyan')
    ax[2].scatter(n, err_rel, c='skyblue')
ax[0].loglog()
ax[0].set_ylabel('mean bias', fontsize = 12)
ax[1].set_ylabel('mean error on signal', fontsize = 12)
ax[0].set_xlabel('number of bins', fontsize = 12)
ax[1].set_xlabel('number of bins', fontsize = 12)
```

```python
x_low, x_up = 1, 10
def error_list(n, sigma_signal, ns):
    x_corner = np.logspace(np.log10(x_low), np.log10(x_up), n + 1)
    mean_x_corner = np.array([np.mean([x_corner[i], x_corner[i+1]]) for i in range(len(x_corner)-1)])
    y_binned = np.zeros(len(x_corner)-1)
    #binned vector
    for i in range(len(x_corner)-1):
        y_binned[i] = binng_model(x_corner[i], x_corner[i+1], a)
    y_true_center = model(mean_x_corner, a)
    #error:
    S = np.zeros(len(x_corner)-1)
    for i in range(len(x_corner)-1):
        S[i] = quad(dV_dx, x_corner[i], x_corner[i+1])[0]
    err = sigma_signal/np.sqrt(ns * S)
    return err, y_binned, y_true_center
```

```python
def f(p, sigma_signal, ns):
    n, lambda_0 = p
    n = round(n)
    # bias derivative
    err_plus, y_binned_plus, y_true_center_plus = error_list(n + 1, sigma_signal, ns)
    err, y_binned, y_true_center = error_list(n, sigma_signal, ns)
    bias_plus= np.mean((1-y_binned_plus/y_true_center_plus))
    bias = np.mean((1-y_binned/y_true_center))
    deriv_b = bias_plus-bias
    #err relatives derivatives
    sigma_plus = np.mean(err_plus/y_binned_plus)
    sigma = np.mean(err/y_binned)
    deriv_sigma = sigma_plus-sigma
    first = deriv_b - lambda_0*deriv_sigma
    second = sigma - 1
    return first, second
```
