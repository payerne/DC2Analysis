```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def mean(x, theta):
    a,b = theta
    return np.exp(-a*x + -b*x**2)
def mean_curvefit(x, a, b):
    theta = a,b
    return mean(x, theta)
```

```python
x = np.linspace(-np.pi,np.pi,10)
a_true, b_true =1,.1
y_mean = mean(x,[a_true, b_true])
std = 0.2
```

```python
y_random = np.random.randn(300, len(x)) 
```

```python
y_sample = y_random * std + y_mean
cov = np.cov(y_sample.T)
fit = []
sigma2 = []
for y in y_sample:
    popt, sigma2_ = curve_fit(mean_curvefit, x, y, sigma = np.sqrt(cov.diagonal()))
    fit.append(np.array(popt))
    sigma2.append(sigma2_.diagonal())
    plt.scatter(x, y, s = 0.1)
fit = np.array(fit)
```

```python
plt.hist(fit[:,1], bins = 20, alpha = 0.5)
plt.hist(fit[:,0], bins = 20, alpha = 0.5)
```

```python
def Fisher_Matrix(theta, model, x, cov):
    zeros = np.zeros(len(theta))
    delta = 1e-3
    derivative = []
    for i in range(len(theta)):
        delta_i = zeros
        delta_i[i] = delta
        derivative.append((model(x, np.array(theta)) - model(x, np.array(theta) + delta_i))/delta)
    Fisher_matrix = np.zeros([len(theta), len(theta)])
    for i in range(len(theta)):
        for j in range(len(theta)):
            Fisher_matrix[i,j] = np.sum(derivative[i]*np.linalg.inv(np.diag(np.diag(cov))).dot(derivative[j]))
    Fisher_matrix_inv = np.linalg.inv(Fisher_matrix)
    Fisher_matrix_diag_inv = np.linalg.inv(np.diag(Fisher_matrix.diagonal()))
    return Fisher_matrix_inv, Fisher_matrix_diag_inv
```

```python
fig, ax = plt.subplots(1,2, figsize = (15,7))
x_a, x_b = [], []
for i, f in enumerate(fit):
    Fisher_matrix_inv, Fisher_matrix_diag_inv = Fisher_Matrix(f, mean, x, cov)
    x_a.append(np.sqrt(Fisher_matrix_inv.diagonal()[0]))
    x_b.append(np.sqrt(Fisher_matrix_inv.diagonal()[1]))
    #ax[0].scatter(np.sqrt(Fisher_matrix_diag_inv.diagonal()[0]), np.sqrt(Fisher_matrix_inv.diagonal())[0])
    ax[0].scatter(np.sqrt(Fisher_matrix_inv.diagonal()[0]), np.sqrt(sigma2[i][0]), c='k')
    ax[1].scatter(np.sqrt(Fisher_matrix_inv.diagonal()[1]), np.sqrt(sigma2[i][1]), c='k')
    
    ax[0].scatter(np.sqrt(Fisher_matrix_diag_inv.diagonal()[0]), np.sqrt(sigma2[i][0]),c='r')
    ax[1].scatter(np.sqrt(Fisher_matrix_diag_inv.diagonal()[1]), np.sqrt(sigma2[i][1]),c='r')



ax[0].plot(x_a,x_a)
ax[1].plot(x_b,x_b)
```
