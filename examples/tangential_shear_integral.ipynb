```python
import numpy as np
from scipy.integrate import quad, simps
from scipy.special import jn
import matplotlib.pyplot as plt
```

```python
def shear(r):
    return np.exp(-r**2)
```

```python
shear(0)
```

```python
def integrand(theta, phi, theta_0, phi_0, l, beta):
    a = np.cos(2*(beta-phi))
    b = np.exp(1j*l*( theta_0*np.cos(phi_0-beta) + theta*np.cos(phi-beta) ) )
    return a*b
```

```python
def integrer(theta, theta_0, phi_0, l, beta):
    def __integrand__(x):
        return -np.real(integrand(theta, x, theta_0, phi_0, l, beta))
    return quad(__integrand__, 0, 2*np.pi)[0]
```

```python
theta = np.linspace(0,10,100)
theta_0 = np.linspace(0,10,5)
phi_0 = np.linspace(0,np.pi*2,5)
l, beta = 1, np.pi/4
for i in range(5):
    a = []
    for theta_ in theta:
        a.append(integrer(theta_, theta_0[i], phi_0[i], l, beta)/(2*np.pi))
    plt.plot(theta, a)
plt.plot(theta, jn(2, l*theta), '+')
```
