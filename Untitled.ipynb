```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def psosie(Pch, Ntot):
    n =np.arange(1,Ntot + 1)
    pn = (Pch**n)*(1 - Pch)**(Ntot-n)
    norm = np.sum(pn)
    pn_plus_1 = pn
    Psosie = np.sum(pn_plus_1)/norm
    return Psosie, norm
```

```python
np.arange(9)
```

```python
psosie(0.9, 2)
```

```python
Pch = np.logspace(-7, -3, 10)
```

```python
for j, ntot in enumerate([1e4]):
    ptot = []
    for i, pch in enumerate(Pch):
        p, norm = psosie(pch, int(ntot))
        print(norm)
        plt.plot(Pch, (Pch**n)*(1 - Pch)**(ntot-n)/norm)
        ptot.append(p)


    plt.loglog(Pch, ptot, label = f'Psosie', linewidth =3)
plt.legend()
```
